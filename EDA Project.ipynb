# Pie Chart.

```python

df=pd.read_csv('E:\\tips.csv')
if 'day' in df.columns and 'tip' in df.columns:
    # Group the data by 'day' and sum the 'tip' values
    grouped_data = df.groupby('day')['tip'].sum()

    # Plotting the pie chart
    plt.pie(grouped_data, labels=grouped_data.index, autopct='%1.1f%%')
    plt.title('Tips by Day')
    plt.show()
else:
    print("CSV does not contain the required 'day' and 'tip' columns.")

```

```python
sns.displot(df['total_bill'])
```

```python
a=df["total_bill"]
mean=a.mean()
median=np.median(a)
mode=a.mode()
print("Mean:-",mean)
print("Median:-",median)
print("Mode:-",mode)
```

```python
sns.displot(a,hist=False)
plt.axvline(mean,color="r",label="mean")
plt.axvline(median,color="g",label="median")
plt.axvline(mode[0],color="o",label="mode")
plt.legend()
plt.show()
```

```python
sns.boxplot(a)
plt.text(0.85,13,s='A1',size=13)
plt.text(0.85,17,s='A2',size=13)
plt.text(0.85,23,s='A3',size=13)
plt.text(0.1,16,s='1AR',rotation=90,size=13)
plt.show()
```
